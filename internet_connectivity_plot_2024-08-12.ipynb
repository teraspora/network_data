```python
filename = '/golgi/tech24/metrics/logs/ping_8888_2024-07-31.log'    #     ping_8888_2024-08-15.log'
times = []
values = []

with open(filename, 'r') as file:
    for line in file:
        # Step 2: Parse each line
        items = line.strip().split()
        if not items:
            continue
        if items[0].startswith('Ping'):
            server = items[1]
            date = items[3]
            continue
        time_str, value_str = items
        # print(time_str)
        time_obj = time_str[:-3]
        times.append(time_obj)
        values.append(float(value_str))

plt.figure(figsize=(20, 6))
plt.plot(times, values, marker='.')
plt.xlabel('Time of day')
plt.ylabel('Ping time (ms)')
plt.title(f'Internet connectivity on {date} as measured by ping times (in ms) to {server}')
plt.xticks(rotation=90, fontsize=8)
plt.tick_params(axis='x', labelcolor='#357', labelsize='4')
plt.tight_layout()
```
